```python
# Modules
import os
import csv
import pandas as pd
```

```python
capital_file = os.path.join("raw/Capital_Costs.csv")
print(capital_file)
```

```python
# Use Pandas to read data
capital_costs_df = pd.read_csv(capital_file)

# remove any spaces in the cell
capital_costs_df.columns = capital_costs_df.columns.to_series().apply(lambda x: x.strip())
capital_costs_df
```

```python
list(capital_costs_df.columns.values)
```

```python
# remove new line
capital_costs_df = capital_costs_df.replace('\n','', regex=True)

# drop columns not related to dog
new_capital_costs_df = capital_costs_df.drop(columns=['Notes', 'Rabbit', 'G. Pig', 'Ferret', 'Sm\nBird', 'Fish'])
list(new_capital_costs_df.columns.values)
```

```python
# rename columns
new_capital_costs_df1 = new_capital_costs_df.rename(index=str, columns={"Costs": "Items", "Sm\nDog": "Sm_Dog", "Med\nDog": "Med_Dog", "Lg Dog": "Lg_Dog", "Cat": "Cat"})
```

```python
new_capital_costs_df1
```

```python
new_capital_costs_df1.dtypes
```

```python
cols = ['Sm_Dog', 'Med_Dog', 'Lg_Dog', 'Cat']

# remove $ sign from cell values
new_capital_costs_df1[cols] = new_capital_costs_df1[cols].replace({'\$': '', ',': ''}, regex=True)

# assign numeric cell values to type float
new_capital_costs_df1[cols] = new_capital_costs_df1[cols].astype(float)

new_capital_costs_df1.dtypes
```

```python
new_capital_costs_df1
```

```python
# export datafame to new_annual_costs.csv in the output folder
export_new_capital_costs_csv = new_capital_costs_df1.to_csv ("data/new_capital_costs.csv", index = None, header=True) 
```
